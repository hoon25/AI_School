## Pytorch Basic

```
import torch
# 0에서 1사이의 랜덤한 숫자 
x = torch.rand(2,3) 
print(x) 
 
# 정규분포에서 샘플링한 값 
x = torch.randn(2,3) 
print(x) 

# 시작과 끝 사이의 랜덤한 자연수 
x = torch.randint(2,5,size = (2,3))
print(x)
```

```
# 0으로 채워진 텐서 
x = torch.zeros(2,3)
print(x) 

# 인자로 들어오는 텐서와 형태가 같은 0로 채워진 텐서 
ref = torch.rand(4,5) 
x = torch.zeros_like(ref)
print(x) 

# 1로 채워진 텐서 
x = torch.ones(2,3) 
print(x) 

# 인자로 들어오는 텐서와 형태가 같은 1로 채워진 텐서 
ref = torch.rand(4,5)
x = torch.ones_like(ref)
print(x)
```

## Perceptron with pytorch 

```
import torch 
import torch.nn as nn # feedforaward를 하기 위한 대부분의 함수를 가지고 있음 
import numpy as np 

x_data = torch.FloatTensor([1.0,2.0])
print(x_data)

linear = nn.Linear(2,1) # 입력으로 들어오는 퍼셉트론 수, 출력으로 나가는 퍼셉트론 수 
print(linear(x_data)) # w1,w2,bias 다 계산한 값, 새롭게 계산할 때 마다 값이 바뀜

sigmoid = nn.Sigmoid() # sigmoid function 계산해 주는 함수

z = sigmoid(linear(x_data)) 
print(z)

## relu 써서 풀어보기 숙제 
# relu = nn.ReLU() 
# nn.LeakyReLu()  # 좀 더 계산이 정확하게 나오는 함수
# z = relu(linear(x_data))
```

```
linear.weight # weight값 나옴
```

```
linear.bias # 가중치의 값이 나옴 
```
